```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
df = pd.read_csv('medical_dataset.csv')
df.head(10)
```

```python
df.describe()
```

```python
output_dir = 'plots'
os.makedirs(output_dir, exist_ok=True)
for column in df.columns:
    plt.figure(figsize=(8, 6))
    if df[column].dtype == 'object' or len(df[column].unique()) < 10:
        sns.countplot(data=df, x=column)
    else:
        sns.histplot(data=df, x=column, kde=True)

    plt.title(f'Distribution of {column}')
    plt.savefig(os.path.join(output_dir, f'{column}_distribution.png'))
    plt.close()
```

```python
print("Dataset size:")
print(len(df))
print("Missing values:")
print(df.isnull().sum())
```

```python

print("\nAge group distribution:")
df['age_group'] = pd.cut(df['age'], bins=[18, 30, 50, 70, 100], labels=['18-29', '30-49', '50-69', '70+'])
print(df['age_group'].value_counts().sort_index())
print("\nGender distribution:")
print(df['sex'].value_counts())
```
